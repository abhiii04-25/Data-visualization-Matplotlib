# Matrixplots

## Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
tipsco=tips[['total_bill','tip','size']]
tipsco
```

```python
tipsco.corr()
```

```python
sns.heatmap(tipsco.corr(),annot=True,)
```

## Cluster map

```python
sns.clustermap(tipsco.corr(),annot=True, cmap='coolwarm')
```

## Pivot table Heat map

```python
pviflights=flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
pviflights
```

```python
sns.heatmap(pviflights)
```
