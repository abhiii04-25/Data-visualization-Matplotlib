# DIstribution Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")
df
```

```python
df['size'].unique()
```

```python
#This is the first plot histogram plot
sns.histplot(df['total_bill'],bins=20,kde=True)
```

```python
#Subplots in histogram
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

## Joining the plots

```python
#Joining the two plots in one plot
sns.jointplot(x="total_bill",y="tip",data=df,kind="scatter")
```

```python
#KDE plot
sns.jointplot(x="total_bill",y="tip",data=df,kind="kde")
```

```python
#HEX plot
sns.jointplot(x="total_bill",y="tip",data=df,kind="hex")
```

```python
#regression plot
sns.jointplot(x="total_bill",y="tip",data=df,kind="reg")
```

## Pairplot

```python
sns.pairplot(df,hue="size",palette="rainbow")
```

## Rugplot

```python
sns.rugplot(df)
```
