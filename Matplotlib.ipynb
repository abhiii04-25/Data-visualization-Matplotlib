# Matplotlib basics

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#create a simple plot
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
#plotting and labeling the plot
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

## Subplot in Matplot

```python
#subplots
plt.subplot(2,2,1)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y**2)
```

## Object oriented way

```python
fig=plt.figure(figsize=(20,8), dpi=100)
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
plt.plot(x,y)
axis2=fig.add_axes([0.2,0.5,0.4,0.3])
plt.plot(x,y)
```

## Types of plots

```python
#scatter plot points
plt.scatter(x,y)
```

```python
#histogram
from random import sample
data=sample(range(1,1000),100) 
plt.hist(data)
```

```python
#boxplot
plt.boxplot(data)
```

## Saving plot

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,.8,.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```

## Working with images

```python
import matplotlib.image as mpimg
```

```python
img=mpimg.imread("cat.jpg")
plt.imshow(img)
plt.axis("off")
```

```python
#cropping the image
cropped_img=img[100:400,200:500]
plt.imshow(cropped_img)
plt.axis("off")
```
