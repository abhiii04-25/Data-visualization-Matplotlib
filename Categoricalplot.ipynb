# Categorical Data plot 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

## Countplot

```python
#Countplot to analyze the distribution of categorical data
sns.countplot(x=df["sex"],hue=df["smoker"])
```

## Barplot

```python
#Barplot to analyze the distribution of categorical data
sns.barplot(x=df["sex"],y=df["total_bill"])
```

```python
import numpy as np
sns.barplot(x=df["sex"],y=df["total_bill"],estimator=np.sum)
```

## Boxplot

```python
#Barplot for analyzing the distribution of categorical data of tips by day of the week
sns.boxplot(x="tip",y="day",data=df,palette="rainbow",hue="sex")
```

## Violinplot

```python
#Violinplot for analyzing the distribution of categorical data of day and total bill
sns.violinplot(x="day",y="total_bill",data=df,palette="rainbow",hue="sex")
```

## Stripplot and swarmplot

```python
#Stripplot for analyzing the distribution of categorical data of day and total bill
sns.stripplot(x="day",y="total_bill",data=df,palette="rainbow",hue="sex")
```

```python
#Swarmplot for analyzing the distribution of categorical data of day and total bill
sns.swarmplot(x="day",y="total_bill",data=df,palette="rainbow",hue="sex")
```

```python
#Merging the swarmplot and violinplot 
sns.violinplot(x="tip",y="day",data=df,palette="rainbow",hue="sex")
sns.swarmplot(x="tip",y="day",data=df,palette="rainbow",hue="sex")
```
